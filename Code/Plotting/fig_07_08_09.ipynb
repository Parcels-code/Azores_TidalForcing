```python
labels_all = ['01/07/2009', '15/07/2009', '01/08/2009', '15/08/2009', '01/09/2009', '15/09/2009', '01/10/2009', '15/10/2009', '01/11/2009', '15/11/2009', '01/12/2009', '15/12/2009', '01/01/2010', '15/01/2010', '01/02/2010', '15/02/2010','01/03/2010', '15/03/2010', '01/04/2010', '15/04/2010',  '01/05/2010', '15/05/2010', '01/06/2010', '15/06/2010']
x = np.arange(len(labels_all))  # the label locations
```

## Line plot:

```python
fig = plt.figure(figsize=(16, 12)) 

ax1 = plt.subplot(211)
    
# FTLE backward        
var_nT_a = []
var_wT_a = []

for ii in range(0, len(labels_all)):
    
    FTLE_b_nT = dFTLE_nTs[ii]['FTLE_b']
    FTLE_b_wT = dFTLE_wTs[ii]['FTLE_b']
    
    var_nT_a.append(((np.count_nonzero(FTLE_b_nT >= 0.5)) / nb_points) * 100)
    var_wT_a.append(((np.count_nonzero(FTLE_b_wT >= 0.5)) / nb_points) * 100)

rects1 = ax1.plot(x, var_nT_a, '.-', color='C0', linewidth=3, markersize=15, label='No tidal forcing')
rects1 = ax1.plot(x, var_wT_a, '.-', color='C3', linewidth=3, markersize=15, label='Tidal forcing')

ax1.legend(fontsize=16)
        
ax1.set_xticks(x)
ax1.set_xticklabels([])
ax1.grid() 
ax1.set_ylabel('Percentage bFTLE >= 0.5', size=16)

# Skewness
skew_nT_a = []
skew_wT_a = []

ax1 = plt.subplot(212)    
    
for ii in range(0, len(labels_all)):
    
    FTLE_b_nT = dFTLE_nTs[ii]['FTLE_b']
    FTLE_b_wT = dFTLE_wTs[ii]['FTLE_b']
    
    skew_nT_a.append(scipy.stats.skew(FTLE_b_nT.ravel(), nan_policy='omit'))
    skew_wT_a.append(scipy.stats.skew(FTLE_b_wT.ravel(), nan_policy='omit'))
    
rects1 = ax1.plot(x, skew_nT_a, '.-', color='C0', linewidth=3, markersize=15)
rects2 = ax1.plot(x, skew_wT_a, '.-', color='C3', linewidth=3, markersize=15)
        
ax1.set_xticks(x)
ax1.set_xticklabels(labels_all, rotation=45, horizontalalignment='right')
ax1.grid()
ax1.set_ylabel('Skewness', size=16)
ax1.set_xlabel('\n Date [dd/MM/YYYY]', size=16)

plt.savefig('../../Figures/' + 'FTLE_backward_v02_biweekly_perc05_SKEW_nT_wT_AllMonths_newSkew_LinePlot.jpg', dpi=300)
plt.show()
```
