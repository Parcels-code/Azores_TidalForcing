```python
bins_list = np.arange(0, 1.05, 0.05)
skew_nT_1 = []
skew_wT_1 = []

fig = plt.figure(figsize=(16, 10)) 
gs = gridspec.GridSpec(3, 4)

ntitles = nmonths
    
for ii in range(0, len(ntitles)):
    
    FTLE_b_nT_w01 = dFTLE_nTs_w01[ii]['FTLE_b']
    FTLE_b_wT_w01 = dFTLE_wTs_w01[ii]['FTLE_b']
    
    counts_nT, binEdges_nT = np.histogram(FTLE_b_nT_w01.ravel(), bins=bins_list, density=True)
    counts_wT, binEdges_wT = np.histogram(FTLE_b_wT_w01.ravel(), bins=bins_list, density=True)

    skew_nT = stats.skew(counts_nT.ravel())
    skew_wT = stats.skew(counts_wT.ravel())
    skew_nT_1.append(skew_nT)
    skew_wT_1.append(skew_wT)
    
    ###
    print(ii)
    print(counts_nT[-1])
    print(counts_wT[-1])
    print('####')
    ###
    
    bincenters_nT = 0.5 * (binEdges_nT[1:] + binEdges_nT[:-1])
    bincenters_wT = 0.5 * (binEdges_wT[1:] + binEdges_wT[:-1])
        
    ax1 = plt.subplot(gs[ii])    
    
    ax1.plot(bincenters_nT, counts_nT, '.-', color='C0', label='No tidal forcing', linewidth=2, markersize=10)# , linewidth=4
    ax1.plot(bincenters_wT, counts_wT, '.-', color='C3', label='Tidal forcing', linewidth=2, markersize=10)
    
    ax1.grid()
    ax1.tick_params(axis='both', which='major', labelsize=12)
    ax1.tick_params(axis='both', which='minor', labelsize=10)

    ax1.set_ylim([0, 4.5])
    ax1.set_xlim([0, 1.1])
    
    if ii == 0:
        legend = ax1.legend(loc=0, fontsize=12)
        legend.get_frame().set_alpha(None)
        legend.get_frame().set_facecolor('white')
    if ((ii==4)):
        ax1.set_ylabel('Normalized frequency', size=20) # Normalized bin counts
    if ((ii==9)):
        ax1.set_xlabel('                                    Backward FTLE [1/days]', size=20)
    if ii<8 :
        ax1.axes.xaxis.set_ticklabels([])
    if (ii != 0) & (ii != 4) & (ii != 8) :
        ax1.axes.yaxis.set_ticklabels([])
    
    ax1.set_title(ntitles[ii], size=16)    
    ax1.axvline(x=0.5, color='k') 
    
plt.savefig('../../Figures/' + 'FTLE_backward_v02_biweekly_AllMonths_day01_hist1d_bins1_vline05.jpg', dpi=300)

plt.show()
```
